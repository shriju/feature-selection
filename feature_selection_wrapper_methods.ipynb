### Regression

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df['medv'], test_size = 0.2, random_state= 1)

print(X_train.shape)
print(X_test.shape)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# using a plain model without feature selection
import numpy as np
from sklearn.metrics import r2_score
model = LinearRegression()
```

```python
print("training: ", np.mean(cross_val_score(model, X_train, y_train, cv = 5, scoring='r2')))
print("testing: ", np.mean(cross_val_score(model, X_test, y_test, cv = 5, scoring = 'r2' )))
```

```python
# applying feature selection
```

### Exhaustive Feature Selection

```python
lr = LinearRegression()
```

```python
exh = EFS(lr, max_features=12, scoring='r2', cv = 10, print_progress=True, n_jobs= -1)

sel = exh.fit(X_train, y_train)
```

```python
sel.best_score_
```

```python
sel.best_feature_names_
```

```python
metric_df = pd.DataFrame.from_dict(sel.get_metric_dict()).T
metric_df
```

```python
#calculating adjusted r2 score as r2 score is not that reliable for high dimentional data
def adj_r2(r2, num_examples, num_features):
  coef = (num_examples -1)/(num_examples-num_features - 1)
  return 1 - (1-r2) * coef
```

```python
metric_df['observations'] = 404
metric_df['num_features'] = metric_df['feature_idx'].apply(lambda x:len(x))
metric_df['adjusted_r2'] = adj_r2(metric_df['avg_score'],metric_df['observations'],metric_df['num_features'])
```

```python
metric_df.sort_values('adjusted_r2',ascending=False)
```

```python
X_train_sel = sel.transform(X_train)
X_test_sel = sel.transform(X_test)
```

```python
model = LinearRegression()

print("training",np.mean(cross_val_score(model, X_train_sel, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test_sel, y_test, cv=5, scoring='r2')))
```

### Backward Elimination/ Sequential Backward Selection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# load the dataset
data= pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")

#seperate the target variables
X = data.drop("medv", axis =1)
y= data['medv']

#split the data into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

print(X_train.shape)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
model = LinearRegression()

print("Training: ", np.mean(cross_val_score(model, X_train, y_train, cv = 5, scoring = 'r2')))
print("Testing: ", np.mean(cross_val_score(model, X_test, y_test, cv = 5, scoring = 'r2')))
```

```python
lr = LinearRegression()

# perform backward elimination
sfs = SFS(lr, k_features='best', forward=False, floating=False, scoring ='r2', cv=5 )
sfs.fit(X_train, y_train)
```

```python
sfs.k_feature_idx_
```

```python
metric_df = pd.DataFrame.from_dict(sfs.get_metric_dict()).T

metric_df['observations'] = 404
metric_df['num_features'] = metric_df['feature_idx'].apply(lambda x:len(x))
metric_df['adjusted_r2'] = adj_r2(metric_df['avg_score'],metric_df['observations'],metric_df['num_features'])

metric_df
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

fig1 = plot_sfs(sfs.get_metric_dict(), kind = 'std_err')

plt.title("'Sequential Backward Elimination with StdErr")
plt.grid()
plt.show()
```

```python
X_train_sel = sfs.transform(X_train)
X_test_sel = sfs.transform(X_test)

model = LinearRegression()

print("training",np.mean(cross_val_score(model, X_train_sel, y_train, cv=5, scoring='r2')))
print("testing",np.mean(cross_val_score(model, X_test_sel, y_test, cv=5, scoring='r2')))
```
